```python
from playwright_computer_use.async_api import PlaywrightComputerTool
from playwright.async_api import async_playwright, Playwright
```

```python
async with async_playwright() as playwright:
    browser = await playwright.firefox.launch(headless=False)
    context = await browser.new_context()
    page = await context.new_page()
    await page.set_viewport_size({"width": 1024, "height": 768})  # Computer-use default
    await page.goto("https://www.google.com")
    playwright_tools = PlaywrightComputerTool(page, use_cursor=True)
    out = await playwright_tools(action="screenshot")
    print(out)
```
